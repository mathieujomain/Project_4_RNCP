```python
col_list = ['ID', 'URL', 'Titre', 'Chapeau', 'Description', 'Date de début',
       'Date de fin', 'Occurrences', 'Description de la date',
       "URL de l'image", "Texte alternatif de l'image", "Crédit de l'image",
       'Mots clés', 'Nom du lieu', 'Adresse du lieu', 'Code postal', 'Ville',
       'Coordonnées géographiques', 'Accès PMR', 'Accès mal voyant',
       'Accès mal entendant', 'Transport', 'Url de contact',
       'Téléphone de contact', 'Email de contact', 'URL Facebook associée',
       'URL Twitter associée', 'Type de prix', 'Détail du prix',
       "Type d'accès", 'URL de réservation', 'URL de réservation - Texte',
       'Date de mise à jour', 'Image de couverture', 'Programmes',
       'En ligne - address_url', 'En ligne - address_url_text',
       'En ligne - address_text', 'title_event']

#At firt, I've set 876 missing values as a treshold. But after a quick look of the columns deleted, I realised that columns with contact email
# and contact phone number have been drop. So I made a change (969) to be able to keep those columns even if they have a lot of missing values.

def drop_column(data, list):
        for i in list:
            if data[i].isna().sum() > 969:
                del data[i]

drop_column(df, col_list)
```

```python
# We will continue to drop column that we identified as not relevant
drop_col_list = ['Description', 'Description de la date', "URL de l'image",
                 "Texte alternatif de l'image", "Crédit de l'image", 'Coordonnées géographiques',
                 'Détail du prix', 'Image de couverture', 'title_event']

for i in drop_col_list:
    del df[i]
```

```python
#lower content of some columns
lower_list = ['Titre', 'Chapeau', 'Mots clés', 'Nom du lieu', 'Adresse du lieu', 'Ville']

def lower_content(data, list):
        for i in list:
            data[i] = data[i].str.lower()

lower_content(df, lower_list)
```

```python
#Have a look of the data in column Mots clés
df['Mots clés'].value_counts()
```

```python
#Making some changes in Mots clés to only keep one tag per row

activities_list = ['musique', 'conférence', 'expo', 'sport', 'théâtre', 'enfants',
                   'balade', 'loisirs', 'art', 'atelier', 'cinéma', 'spectacle musical',
                   'littérature', 'clubbing', 'cirque', 'nature', 'concert']

#function to replace activities by a single given activity
def keywords(alist):
    for activity in alist:
        s='.*(?='+str(activity)+').*'
       # print(s)  this was to print my paterns to check if it's working
        df['Mots clés'] = np.where(df['Mots clés'].str.contains(activity), df['Mots clés'].str.replace(s, activity, regex=True), df['Mots clés'])
    
keywords(activities_list)
```

```python
#clean list of values
df['Mots clés'].value_counts()
```

```python
#Working on the content of column Date de début and Date de fin
df['Date de début'] = df['Date de début'].str.findall(r'[0-9]{4}-[0-9]{2}-[0-9]{2}')
```

```python
#It has created a list of one value
df['Date de début']
```

```python
#Dropping column with date as list + create new column date
df1 = df.apply(lambda x: pd.Series(df['Date de début'][0]),axis=1).stack().reset_index(level=1, drop=True)
df1.name = 'Date debut'
df=df.drop('Date de début', axis=1).join(df1)
```

```python
#Same process with Date de fin
df['Date de fin'] = df['Date de fin'].str.findall(r'[0-9]{4}-[0-9]{2}-[0-9]{2}')
df2 = df.apply(lambda x: pd.Series(df['Date de fin'][0]),axis=1).stack().reset_index(level=1, drop=True)
df2.name = 'Date fin'
df=df.drop('Date de fin', axis=1).join(df2)
```

```python
#Same process with Date de mise à jour
df['Date de mise à jour'] = df['Date de mise à jour'].str.findall(r'[0-9]{4}-[0-9]{2}-[0-9]{2}')
df3 = df.apply(lambda x: pd.Series(df['Date de mise à jour'][0]),axis=1).stack().reset_index(level=1, drop=True)
df3.name = 'Date mise a jour'
df=df.drop('Date de mise à jour', axis=1).join(df3)
```

```python
#Checking phone number column
df['Téléphone de contact'].value_counts()
```

```python
#We need to replace spaces and dots by nothing
df['Téléphone de contact'] = df['Téléphone de contact'].str.replace(' ','',regex=True)
df['Téléphone de contact'] = df['Téléphone de contact'].str.replace('.','',regex=True)
```

```python
#Checking postal code column
df['Code postal'].value_counts()
```

```python
#We need to do some replacement
df['Code postal'] = df['Code postal'].str.replace(' ','',regex=True)
df['Code postal'] = np.where(df['Code postal'].str.contains('75003P'), df['Code postal'].str.replace('75003P', '75003', regex=True), df['Code postal'])
df['Code postal'] = np.where(df['Code postal'].str.contains('75002Paris'), df['Code postal'].str.replace('75002Paris', '75002', regex=True), df['Code postal'])
df['Code postal'] = np.where(df['Code postal'].str.contains('75011/75012'), df['Code postal'].str.replace('75011/75012', '75011', regex=True), df['Code postal'])
df['Code postal'] = np.where(df['Code postal'].str.contains('75°16'), df['Code postal'].str.replace('75°16', '75016', regex=True), df['Code postal'])
```

```python
#Replacing nan values or not valid value by postal code 75000 (standard for Paris)
df['Code postal'] = np.where(df['Code postal'].str.contains('4e/11e/12e'), df['Code postal'].str.replace('4e/11e/12e', '75000', regex=True), df['Code postal'])
df['Code postal'] = np.where(df['Code postal'].str.contains('Paris'), df['Code postal'].str.replace('Paris', '75000', regex=True), df['Code postal'])
df['Code postal']=df['Code postal'].replace(np.nan, '75000')
```

```python
#Making sure it is all good
df['Code postal'].isnull().sum()
```

```python
#Do some changes in the Occurrences column to actually have a count instead of dates
df['Occurrences'] = df['Occurrences'].str.findall(r'[0-9]{4}-[0-9]{2}-[0-9]{2}')
```

```python
df['Occurrences']=df['Occurrences'].replace(np.nan, 0)
```

```python
df['Nb occurrences'] = None
for i, k in zip(df['Occurrences'], range(0, len(df['Occurrences']))):
        if i != 0:
            x=len(i)
            df['Nb occurrences'][k]=x

df.drop('Occurrences', axis=1)
```

```python
#I do think we need to take care of values
df['Nb occurrences'].max()
```

```python
#Replace delete value with more than 100
df.loc[df['Nb occurrences'] > 100, 'Nb occurrences'] = np.nan
```

```python
#We need to replace none data by a text that say so in some column
#First step check again nan values
df.isna().sum()
```

```python
#Replace missing values in columns by non renseigné
col_missing_values = ['Chapeau', 'Mots clés', 'Nom du lieu', 'Adresse du lieu',
                      'Ville', 'Url de contact', 'Téléphone de contact',
                      'Email de contact', 'URL Facebook associée', 'Type de prix',
                      "Type d'accès", 'URL de réservation']

def replace_values(col_list):
    for i in col_list:
        df.loc[(df[i].isnull()), i] = 'non renseigne'

replace_values(col_missing_values)
```

```python
#Checking result
df.isna().sum()
```

```python
#Another verification
df['Ville'].value_counts()
```

```python
#Seems we still have some work to do in the column Ville
df.loc[(df['Ville'].str.contains('paris')), 'Ville'] = 'paris'
df.loc[(df['Ville'].str.contains('parís')), 'Ville'] = 'paris'
df.loc[(df['Ville'].str.contains('75001')), 'Ville'] = 'paris'
```

```python
#In this step we are going to change every letter with an accent by the normal letter. That way we won't have problem with sql afterwards

accent_col_list = ['Titre', 'Chapeau', 'Mots clés', 'Nom du lieu',
                   'Adresse du lieu', 'Ville', "Type d'accès"]

df[accent_col_list] = df[accent_col_list].apply(lambda x: x.str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8'))
```
